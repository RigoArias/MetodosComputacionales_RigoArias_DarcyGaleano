```python
#Ejercicios derivadas 8
#ejercicio E

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.sqrt(np.tan(x))

def derivada_exacta(x):
  return ((1 / np.cos(x))**2)/(2*np.sqrt(np.tan(x)))

def derivada_progresiva(f,x,h):
  return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))

def derivada_central(f,x,h):
  return (f(x+h)-f(x-h))/(2*h)

h=0.01
intervalo= np.arange(0.1,1.1,h)

derivadas_progresivas = [derivada_progresiva(f, x, h) for x in intervalo]
derivadas_centrales= [derivada_central(f, x, h) for x in intervalo]
derivada_analitica= derivada_exacta(intervalo)


plt.figure(figsize=(12, 8))

plt.plot(intervalo, derivada_analitica, label='Derivada Analítica', color='black', marker= "x")
plt.plot(intervalo, derivadas_centrales, label='Derivada Central Estimada', linestyle='--', color='blue')
plt.plot(intervalo, derivadas_progresivas, label='Derivada Progresiva Estimada', linestyle='-.', color='red')

plt.title("Comparación de Derivadas Analítica y Aproximadas")
plt.legend()
plt.show()




```

```python
#Ejercicios derivadas 8
#ejercicio F

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.sqrt(np.tan(x))

def derivada_exacta(x):
  return ((1 / np.cos(x))**2)/(2*np.sqrt(np.tan(x)))

def derivada_progresiva(f,x,h):
  return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))

def derivada_central(f,x,h):
  return (f(x+h)-f(x-h))/(2*h)

h = 0.01
intervalo = np.arange(0.1, 1.1, h)

derivadas_centrales = [derivada_central(f, x, h) for x in intervalo]
derivadas_progresivas = [derivada_progresiva(f, x, h) for x in intervalo]
derivada_analitica = derivada_exacta(intervalo)

error_central = np.abs(np.array(derivadas_centrales) - derivada_analitica)
error_progresiva = np.abs(np.array(derivadas_progresivas) - derivada_analitica)

plt.plot(intervalo, error_central, label="Error Nodal Central", linestyle='--', color='green')
plt.plot(intervalo, error_progresiva, label="Error Nodal Progresivo", linestyle='-.', color='red')
plt.title("Comparación del Error Nodal para las Aproximaciones Central y Progresiva O(h^2)")
plt.legend()

plt.show()

#Con la grafica podemos notar que si tienen el mismo orden de precision
```
